```python
# ==========================================================
# Experiment: GAN for CIFAR-10 Image Generation using PyTorch
# Aim: Implement a GAN to generate realistic images using CIFAR-10
# ==========================================================

# Install if needed:
# pip install torch torchvision matplotlib

import torch
import torch.nn as nn
import torch.optim as optim
import torchvision
import torchvision.transforms as transforms
from torchvision.utils import save_image, make_grid
from torch.utils.data import DataLoader
import matplotlib.pyplot as plt
import os

# ==========================================================
# 1. Device Configuration
# ==========================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using Device:", device)

# ==========================================================
# 2. Hyperparameters
# ==========================================================
batch_size = 128
lr = 0.0002
latent_dim = 100
epochs = 50
image_size = 32
channels = 3

# ==========================================================
# 3. Create Folder for Saving Images
# ==========================================================
os.makedirs("generated_images", exist_ok=True)

# ==========================================================
# 4. Load CIFAR-10 Dataset
# ==========================================================
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5, 0.5, 0.5),
                         (0.5, 0.5, 0.5))   # Range [-1,1]
])

dataset = torchvision.datasets.CIFAR10(
    root="./data",
    train=True,
    download=True,
    transform=transform
)

loader = DataLoader(dataset, batch_size=batch_size, shuffle=True)

# ==========================================================
# 5. Generator Model
# ==========================================================
class Generator(nn.Module):
    def __init__(self):
        super(Generator, self).__init__()

        self.model = nn.Sequential(
            nn.ConvTranspose2d(latent_dim, 256, 4, 1, 0, bias=False),
            nn.BatchNorm2d(256),
            nn.ReLU(True),

            nn.ConvTranspose2d(256, 128, 4, 2, 1, bias=False),
            nn.BatchNorm2d(128),
            nn.ReLU(True),

            nn.ConvTranspose2d(128, 64, 4, 2, 1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(True),

            nn.ConvTranspose2d(64, 3, 4, 2, 1, bias=False),
            nn.Tanh()
        )

    def forward(self, x):
        return self.model(x)

# ==========================================================
# 6. Discriminator Model
# ==========================================================
class Discriminator(nn.Module):
    def __init__(self):
        super(Discriminator, self).__init__()

        self.model = nn.Sequential(
            nn.Conv2d(3, 64, 4, 2, 1),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(64, 128, 4, 2, 1),
            nn.BatchNorm2d(128),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Conv2d(128, 256, 4, 2, 1),
            nn.BatchNorm2d(256),
            nn.LeakyReLU(0.2, inplace=True),

            nn.Flatten(),
            nn.Linear(256 * 4 * 4, 1),
            nn.Sigmoid()
        )

    def forward(self, x):
        return self.model(x)

# ==========================================================
# 7. Initialize Models
# ==========================================================
G = Generator().to(device)
D = Discriminator().to(device)

# ==========================================================
# 8. Loss Function and Optimizers
# ==========================================================
criterion = nn.BCELoss()

optimizer_G = optim.Adam(G.parameters(), lr=lr, betas=(0.5, 0.999))
optimizer_D = optim.Adam(D.parameters(), lr=lr, betas=(0.5, 0.999))

# ==========================================================
# 9. Training GAN
# ==========================================================
for epoch in range(epochs):

    for i, (real_images, _) in enumerate(loader):

        real_images = real_images.to(device)
        batch = real_images.size(0)

        real_labels = torch.ones(batch, 1).to(device)
        fake_labels = torch.zeros(batch, 1).to(device)

        # -----------------------------------
        # Train Discriminator
        # -----------------------------------
        optimizer_D.zero_grad()

        outputs_real = D(real_images)
        loss_real = criterion(outputs_real, real_labels)

        noise = torch.randn(batch, latent_dim, 1, 1).to(device)
        fake_images = G(noise)

        outputs_fake = D(fake_images.detach())
        loss_fake = criterion(outputs_fake, fake_labels)

        loss_D = loss_real + loss_fake
        loss_D.backward()
        optimizer_D.step()

        # -----------------------------------
        # Train Generator
        # -----------------------------------
        optimizer_G.zero_grad()

        outputs = D(fake_images)
        loss_G = criterion(outputs, real_labels)

        loss_G.backward()
        optimizer_G.step()

        # -----------------------------------
        # Print Progress
        # -----------------------------------
        if i % 100 == 0:
            print(f"Epoch [{epoch+1}/{epochs}] "
                  f"Batch [{i}/{len(loader)}] "
                  f"Loss D: {loss_D.item():.4f}, "
                  f"Loss G: {loss_G.item():.4f}")

    # Save generated images after every epoch
    with torch.no_grad():
        sample_noise = torch.randn(64, latent_dim, 1, 1).to(device)
        fake = G(sample_noise)
        save_image(fake, f"generated_images/epoch_{epoch+1}.png",
                   normalize=True)

# ==========================================================
# 10. Generate Final Images
# ==========================================================
with torch.no_grad():
    noise = torch.randn(64, latent_dim, 1, 1).to(device)
    fake_images = G(noise).cpu()

grid = make_grid(fake_images, nrow=8, normalize=True)

plt.figure(figsize=(8,8))
plt.imshow(grid.permute(1,2,0))
plt.title("Generated CIFAR-10 Images")
plt.axis("off")
plt.show()
```

```python
# ===============================================================
# Experiment No. 08
# Recurrent Neural Network (RNN) using PyTorch
# Aim: Predict sequential data using a pre-existing dataset
# Dataset Used: Airline Passengers Dataset
# ===============================================================

# Install if needed:
# pip install torch pandas matplotlib scikit-learn

import torch
import torch.nn as nn
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler

# ===============================================================
# 1. Device Configuration
# ===============================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using Device:", device)

# ===============================================================
# 2. Load Pre-existing Dataset
# Airline passengers data (monthly totals)
# ===============================================================
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/airline-passengers.csv"
df = pd.read_csv(url)

print(df.head())

# Use passenger column only
data = df["Passengers"].values.reshape(-1, 1)

# ===============================================================
# 3. Normalize Data
# ===============================================================
scaler = MinMaxScaler(feature_range=(0, 1))
data_scaled = scaler.fit_transform(data)

# ===============================================================
# 4. Create Sequences
# Use previous 12 months to predict next month
# ===============================================================
seq_length = 12

X = []
Y = []

for i in range(len(data_scaled) - seq_length):
    X.append(data_scaled[i:i+seq_length])
    Y.append(data_scaled[i+seq_length])

X = np.array(X)
Y = np.array(Y)

# Convert to tensors
X = torch.tensor(X, dtype=torch.float32)
Y = torch.tensor(Y, dtype=torch.float32)

# ===============================================================
# 5. Train-Test Split
# ===============================================================
train_size = int(len(X) * 0.8)

X_train = X[:train_size].to(device)
Y_train = Y[:train_size].to(device)

X_test = X[train_size:].to(device)
Y_test = Y[train_size:].to(device)

# ===============================================================
# 6. Define RNN Model
# ===============================================================
class RNNModel(nn.Module):
    def __init__(self):
        super(RNNModel, self).__init__()

        self.rnn = nn.RNN(
            input_size=1,
            hidden_size=64,
            num_layers=1,
            batch_first=True
        )

        self.fc = nn.Linear(64, 1)

    def forward(self, x):
        h0 = torch.zeros(1, x.size(0), 64).to(device)

        out, _ = self.rnn(x, h0)

        out = out[:, -1, :]   # last time step
        out = self.fc(out)

        return out

# ===============================================================
# 7. Initialize Model
# ===============================================================
model = RNNModel().to(device)

criterion = nn.MSELoss()
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)

# ===============================================================
# 8. Train Model
# ===============================================================
epochs = 200

for epoch in range(epochs):

    model.train()

    output = model(X_train)
    loss = criterion(output, Y_train)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 20 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.6f}")

# ===============================================================
# 9. Evaluate Model
# ===============================================================
model.eval()

with torch.no_grad():
    pred = model(X_test)

# Convert back to original values
predicted = scaler.inverse_transform(pred.cpu().numpy())
actual = scaler.inverse_transform(Y_test.cpu().numpy())

# ===============================================================
# 10. Plot Results
# ===============================================================
plt.figure(figsize=(12,5))
plt.plot(actual, label="Actual Passengers")
plt.plot(predicted, label="Predicted Passengers")
plt.title("Airline Passenger Prediction using RNN")
plt.xlabel("Months")
plt.ylabel("Passengers")
plt.legend()
plt.show()

# ===============================================================
# 11. Predict Next Month
# ===============================================================
sample = X_test[0].unsqueeze(0)

with torch.no_grad():
    next_value = model(sample)

next_value = scaler.inverse_transform(next_value.cpu().numpy())

print("Predicted Next Month Passengers:", int(next_value[0][0]))
```

```python
# ===============================================================
# Experiment No. 09
# LSTM using PyTorch
# Aim: Predict the next word in a sequence of text using LSTM
# ===============================================================

# Install if needed:
# pip install torch numpy

import torch
import torch.nn as nn
import torch.optim as optim
import numpy as np

# ===============================================================
# 1. Device Configuration
# ===============================================================
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using Device:", device)

# ===============================================================
# 2. Load Simple Text Corpus
# ===============================================================
text = """
deep learning is powerful
deep learning is useful
machine learning is powerful
machine learning is useful
artificial intelligence is growing
deep neural networks are useful
"""

# Convert text to lowercase words
words = text.lower().split()

# Unique words
vocab = sorted(set(words))

# Create mappings
word_to_idx = {word: i for i, word in enumerate(vocab)}
idx_to_word = {i: word for word, i in word_to_idx.items()}

vocab_size = len(vocab)

print("Vocabulary:", vocab)

# ===============================================================
# 3. Prepare Sequences
# Example:
# [deep learning is] -> powerful
# ===============================================================
seq_length = 3

input_sequences = []
targets = []

for line in text.strip().split("\n"):
    tokens = line.lower().split()

    for i in range(len(tokens) - seq_length):
        seq = tokens[i:i+seq_length]
        target = tokens[i+seq_length]

        input_sequences.append([word_to_idx[w] for w in seq])
        targets.append(word_to_idx[target])

X = torch.tensor(input_sequences, dtype=torch.long).to(device)
Y = torch.tensor(targets, dtype=torch.long).to(device)

print("Training Samples:", len(X))

# ===============================================================
# 4. Define LSTM Model
# ===============================================================
class LSTMModel(nn.Module):
    def __init__(self, vocab_size, embed_dim=10, hidden_dim=64):
        super(LSTMModel, self).__init__()

        self.embedding = nn.Embedding(vocab_size, embed_dim)

        self.lstm = nn.LSTM(
            input_size=embed_dim,
            hidden_size=hidden_dim,
            num_layers=1,
            batch_first=True
        )

        self.fc = nn.Linear(hidden_dim, vocab_size)

    def forward(self, x):
        x = self.embedding(x)

        out, _ = self.lstm(x)

        out = out[:, -1, :]      # last time step output
        out = self.fc(out)

        return out

# ===============================================================
# 5. Initialize Model
# ===============================================================
model = LSTMModel(vocab_size).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)

# ===============================================================
# 6. Train Model
# ===============================================================
epochs = 500

for epoch in range(epochs):

    model.train()

    output = model(X)
    loss = criterion(output, Y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

    if (epoch + 1) % 50 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss: {loss.item():.4f}")

# ===============================================================
# 7. Predict Next Word Function
# ===============================================================
def predict_next_word(model, text_input):
    model.eval()

    words_input = text_input.lower().split()

    if len(words_input) != seq_length:
        print(f"Enter exactly {seq_length} words.")
        return

    seq = [word_to_idx[word] for word in words_input]
    seq = torch.tensor([seq], dtype=torch.long).to(device)

    with torch.no_grad():
        output = model(seq)
        pred_idx = torch.argmax(output, dim=1).item()

    print("Input Sequence :", text_input)
    print("Predicted Word :", idx_to_word[pred_idx])

# ===============================================================
# 8. Test Predictions
# ===============================================================
predict_next_word(model, "deep learning is")
predict_next_word(model, "machine learning is")
predict_next_word(model, "networks are useful")
```
